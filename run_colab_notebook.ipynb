# Preparar Dataset

Este modelo de PassGAN usa las frecuencias de las contraseñas en formato [Frecuancia][contraseña]. Partiendo de Rockyou.txt se ha calculado el archivo Rockyou_frec.txt :

````python
from google.colab import drive
drive.mount('/content/drive')
````

````python
file_path='/content/drive/Othercomputers/My MacBook Pro/PassGAN_data/rockyou.txt'

import pandas as pd

df=pd.read_csv(file_path,
            delimiter = "\n", 
            header = None, 
            names = ["Passwords"],
            encoding = "ISO-8859-1")

df.head()
````

````python
df_frec=df.value_counts().rename_axis('unique_values').reset_index(name='count')#.to_frame('count')
df_frec=df_frec[['count','unique_values']]
df_frec.head()
````

````python
df_frec.to_csv(file_path.replace(".txt","_frec.txt"),header=None, index=None, sep=' ',columns=["count","unique_values"])
````

# Configurar Dataset

Once created the dataset, you have to edit the config file in *PasswordGAN/CONF/arch1.gin*. In particular, you have to set the variable *setup.dataset_path* with the path to the dataset e.g.:

```
setup.dataset_path = "/home/user/dataset.pickle"
```

Then, you can train the model by using the script *train.py*. The script takes as input the path to the config file e.g.,:

 ```
python train.py CONF/arch1.gin
 ```

The script saves tensorboard logs  and checkpoints inside the folder "./PasswordGAN/HOME/CHECKPOINTS/".
