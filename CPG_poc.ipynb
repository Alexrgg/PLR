# Setup model

```python
from CPG_utils import *
import re


cm_path = 'DATA/MISC/AE_char_map.pickle'
cm, cm_ = load_charmap(cm_path)

model_path = 'DATA/TFHUB_models/AE_based_Noise_a4_ls128_hn4_al8/'

# number of samples
n = 50000
# standard devation 
stddev = .8

xph, prediction = setup_model(model_path, n, stddev)
```

# Conditional Password Guessing (CPG)

Use the variable $s$ in the cell bellow to define a password template. You can use the symbol '\t' to express a *wildcard* (i.e., '$*$' in the paper.)

```python
s = "pa\t\tw\trd1\t\t"

if not EMPTY in s:
    raise Exception("No wildcard in the string: %s" % s)
    
print("Template:", re.subn('\t', '*', s)[0])

print_n = 100
gs = curate_sample(infer(s, xph, prediction, cm, cm_), s)

print("\nRealizations:")
for g in gs[:print_n]:
    print('\t', g)
```
